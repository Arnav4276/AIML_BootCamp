```python
#recurrent
#RNN:- recurrent nerual network
#in RNN the same cell is used again and again across sequence
#each and evry words enter the sequence one by one..


'''

At time step t


current input  = x_t (current token)
previous memory = h_(t-1) (updated memory)
new memory = h_t
output = y_t (positive intent or positive sentiment)



```

```python
positive_words = ["good","great","amazing","excellent","love",]
negative_words = ["bad","terrible","awful","horrible","hate"]

def simple_sentiment(sentence):
  words = sentence.lower().replace(".","").split()

  score = 0

  for word in words:
    if word in positive_words:
      score += 1
    elif word in negative_words:
      score -= 1
  if score > 0:
    return "Positive"
  elif score < 0:
    return "Negative"
  else:
    return "Neutral"

print(simple_sentiment("the movie was good"))
print(simple_sentiment("the movie was not good"))
print(simple_sentiment("the movie was bad"))
```

```python
sentence = "the movie was not good"

tokens = sentence.split()

hidden_state = {
    "topic": None,
    "negation_active": False,
    "sentiment_word": None
}

for token in tokens:
  if token in ["movie","food","product"]:
    hidden_state["topic"] = token

  if token == "not":
    hidden_state["negation_active"] = True

  if token in ["good","bad","terrible","great"]:
    hidden_state["sentiment_word"] = token

  print("Token:", token)
  print("Hidden state:", hidden_state)
  print("-"*179)
```

```python
#LSTM is an upgraded RNN

#LSTM has 4 gates :- forget,cell,input,output...

#LSTM = RNN + Gates + cell gates

#Cell state :- it is the long term state memory path

#Hidden state :- it is the current working memory or the current output that comes in

#Forget gate:- decides which old info is to keep and which info to drop...

#Input gate :- it point to the input gate ( write the usefull info to the memeory )

#Output gate :- selects which memories to expose in the current output..



#LSTM in an analogy:- for a student in high school....

    #forget gate :- cross out the old info which i s no longer required..
    #input gate :- decides what the new info deserves a note...
    #Candidate memory:- the new note that could be written for important aspects...
    #Cell state:- is the notebook
    #Output:- the info that the student has learned...




#Sentence :- The movie was not good

#Token 1: The
#cell state:- sentence started

#Token 2: movie
#cell state:- subject is movie

#Token 3: was
#cell state:- waiting for the description

#Tokne 4: not
#input gate: adds negation_active
#cell state:- nagation is active

#Token: good
#output gate:- exposes the combination of not + good

#Prediction: Negative
```

```python
import tensorflow as tf
from tensorflow.keras import layers

vocab_size = 10000
embedding_dim = 16
max_length = 8

simple_rnn_model = tf.keras.Sequential([
    layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length),
    layers.SimpleRNN(units=32),
    layers.Dense(units=1, activation='sigmoid') # output is betwwen 0 and 1, if the value is close to 1 it means the intent is positive and if the output is close to the 0 it means that the intent is negative..
])

simple_rnn_model.summary()
```

```python
#Weaknesses of RNNs:-

#1. short term memory
        #hidden state

#Exact architecture of LSTM and will code the architecture of LSTM..

#LSTM = RNN + Gates + cell gates

#Cell state:- long term info carrier which pushes important info forward or long term memory
#Hidden state:- summarizes the current info that is coming or short term current memory
#Candidate memory:-
#Forget gate:- decides which old info is to keep and which info to drop
#Input gate:-write the usefull info to the memeory
#Output gate:- elects which memories to expose in the current output

#input gate is the decision maker
#candidate memory is the executioner
```

```python
#I love batch 2 because all the students are very active and intelligent, they can grasp the concepts of LSTM and
#process it properly
#if i ask them about_____they will answer

#LSTM

#Token 1:- I
#Hidden state:- sentence started

#Token 2:- Love
#Hidden state:- subject likes something but unclear
#Forget gate:- NAN
#Input gate:- love-showing intent of liking

#Token 3:- Batch 2
#Hidden state:- subject likes the object (batch2)
#Input gate:- subject showing liking to the batch 2

#Tkone 4:- because
#Hidden state- reason incoming
#input gate:- why subjcet like object info upcoming\
#Forget gate:- NAN

#Token 5:- all

#Token 6:- the

#Token 7:- students
#Hidden state::- info of batch 2(students)
#Forget gate:- the

#Output gate:- most relevant option or output
```

```python
#Architecture of LSTM:- binary classification model

#Layers and shapes

import tensorflow as tf      #mathematical function, tensors manipulating tools
from tensorflow.keras import layers     #keras is a high level API for tensorflow where we can do operations
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.models import Sequential

#Hyperparamteers
vocab_size = 1000   #defines the total no of unique words or unique tokens
embedding_dim = 16  #16 vectors ka list banayega ye funtion (or array) from the vocab_size of 1000
max_length = 8      #total no of words that it selects or the length of the sentence

#Model architecture of LSTM
lstm_model = Sequential(
    [
        #layer 1: Input wording embeddings
        Embedding(
            input_dim=vocab_size,
            output_dim=embedding_dim,
            input_length=max_length, #it will take 8 words, 8 integers
            name="embedding_layer"
        ),

        #layer 2: Sequence processing (RNN base:- with lstm touch)
        LSTM(
            units=32, #it will process 32 internal memory cell info -- it process 8 vectors one by one in a sequential manner
            name="lstm_layer"
        ),

        #layer 3: Output binary classification
        Dense(
            units=1,
            activation="sigmoid",
            name="output_layer"
        ),
    ],
    name="Sequential_LSTM_Classifier",
)

#Verification
lstm_model.summary()
```
